```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio

# Assume the following classes and functions are defined elsewhere in your code:
# from your_module import Parameters_for_Analysis, run_simulation_and_get_outcome_instance

def plot_2d_sensitivity_analysis_initial_pop_vs_evolvability(
    N2_N1_ratios,
    V_A2_V_A1_ratios,
    final_outcomes_matrix,
    subplot_row,
    subplot_col,
    fig
):
    """
    Plots a heatmap for a specific competition type and scaling factor.

    Parameters:
    - N2_N1_ratios (array): Ratios of initial populations (fast-ev./slow-ev.).
    - V_A2_V_A1_ratios (array): Ratios of evolvability (V_A2/V_A1).
    - final_outcomes_matrix (2D array): Simulation outcomes.
    - subplot_row (int): Row index in the subplot grid.
    - subplot_col (int): Column index in the subplot grid.
    - fig (Figure): Plotly figure object to add the heatmap to.
    """
    color_map = {
        "sp 2 (high-genetic-var) wins, sp 1 (low-genetic-var) extinct": 0,
        "sp 1 (low-genetic-var) wins, sp 2 (high-genetic-var) extinct": 1,
        "both extinct": 2,
        "both coexist": 3
    }

    # Convert the final outcomes to numerical values
    numerical_matrix = np.vectorize(color_map.get)(final_outcomes_matrix)

    # Define colors for the heatmap
    colorscale = [
        [0.0, 'red'],       # sp 2 wins, sp 1 extinct
        [1/3, 'blue'],      # sp 1 wins, sp 2 extinct
        [2/3, 'black'],     # both coexist
        [1.0, 'green']      # both extinct
    ]

    # Create the heatmap
    heatmap = go.Heatmap(
        z=numerical_matrix.T,  # Transpose the matrix to align axes correctly
        x=N2_N1_ratios,
        y=V_A2_V_A1_ratios,
        colorscale=colorscale,
        showscale=False,  # We'll add a single color scale later
        hovertemplate='N2/N1: %{x}<br>V_A2/V_A1: %{y}<br>Outcome: %{z}<extra></extra>'
    )

    # Add the heatmap to the specified subplot
    fig.add_trace(heatmap, row=subplot_row, col=subplot_col)

    # Update axes for each subplot
    fig.update_xaxes(
        title_text='Initial Population Ratio (log) [fast-ev./slow-ev.]',
        type='log',
        tickvals=[0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000],
        ticktext=["0.001", "0.01", "0.1", "1", "10", "100", "1000", "10000"],
        row=subplot_row,
        col=subplot_col
    )
    fig.update_yaxes(
        title_text='Evolvability Ratio [fast-ev./slow-ev.] (log)',
        type='log',
        tickvals=[1, 10, 100, 1000],
        ticktext=['1', '10', '100', '1000'],
        row=subplot_row,
        col=subplot_col
    )

def generate_parameters_with_varied_parameters(
    N2_N1_ratios,
    V_A2_V_A1_ratios,
    V_A_scale,
    competition_params,
    base_parameters
):
    """
    Generates a list of Parameters_for_Analysis objects with varied parameters.

    Parameters:
    - N2_N1_ratios (array): Ratios of initial populations (fast-ev./slow-ev.).
    - V_A2_V_A1_ratios (array): Ratios of evolvability (V_A2/V_A1).
    - V_A_scale (float): Scaling factor for V_A1 and V_A2.
    - competition_params (dict): Dictionary containing alpha parameters.
    - base_parameters (dict): Base parameters for the analysis.

    Returns:
    - parameters_list (list): List of Parameters_for_Analysis objects.
    """
    parameters_list = []
    for N2_N1 in N2_N1_ratios:
        for V_A2_V_A1 in V_A2_V_A1_ratios:
            # Scale V_A1 and V_A2 based on the scaling factor
            scaled_V_A1 = V_A_scale * 0.01
            scaled_V_A2 = V_A_scale * 0.01  # Adjust V_A2 based on the ratio

            # Update base_parameters with competition parameters
            updated_base_params = base_parameters.__dict__.copy()
            updated_base_params.update(competition_params)
            updated_base_params['V_A1'] = scaled_V_A1
            updated_base_params['V_A2'] = scaled_V_A2

            # Set initial_population_species_2 based on the ratio
            updated_base_params['initial_population_species_2'] = updated_base_params['initial_population_species_1'] * N2_N1
            updated_base_params['V_A2'] = updated_base_params['V_A1'] * V_A2_V_A1

            # Instantiate Parameters_for_Analysis with updated parameters
            parameters = Parameters_for_Analysis(base_parameters = base_parameters, **updated_base_params)
            parameters_list.append(parameters)
    return parameters_list

def run_sensitivity_analysis(parameters_list):
    """
    Runs the simulation for each set of parameters and collects the outcomes.

    Parameters:
    - parameters_list (list): List of Parameters_for_Analysis objects.

    Returns:
    - final_outcomes_list (list): List of simulation outcome strings.
    """
    final_outcomes_list = []
    for parameters in parameters_list:
        final_outcome_instance = run_simulation_and_get_outcome_instance(parameters)
        final_outcomes_list.append(final_outcome_instance.outcome)
    return final_outcomes_list

def run_multipanel_sensitivity_analysis(
    N2_N1_ratios,
    V_A2_V_A1_ratios,
    scale_factors,
    competition_types,
    base_parameters
):
    """
    Runs the sensitivity analysis across different competition types and scaling factors.

    Parameters:
    - N2_N1_ratios (array): Ratios of initial populations (fast-ev./slow-ev.).
    - V_A2_V_A1_ratios (array): Ratios of evolvability (V_A2/V_A1).
    - scale_factors (list): List of scaling factors for V_A1 and V_A2.
    - competition_types (list): List of dictionaries defining competition types and their alpha parameters.
    - base_parameters (dict): Base parameters for the analysis.

    Returns:
    - final_outcomes_matrices (dict): Nested dictionary with outcomes for each competition type and scaling factor.
    """
    final_outcomes_matrices = {}
    for comp_idx, competition in enumerate(competition_types, start=1):
        comp_name = competition['name']
        comp_params = competition['params']
        final_outcomes_matrices[comp_name] = {}
        for scale_idx, scale in enumerate(scale_factors, start=1):
            print(f"Running sensitivity analysis for Competition: {comp_name}, V_A1 & V_A2 scaled by {scale}...")
            parameters_list = generate_parameters_with_varied_parameters(
                N2_N1_ratios,
                V_A2_V_A1_ratios,
                scale,
                comp_params,
                base_parameters
            )
            final_outcomes_list = run_sensitivity_analysis(parameters_list)
            final_outcomes_matrix = np.array(final_outcomes_list).reshape(len(N2_N1_ratios), len(V_A2_V_A1_ratios))
            final_outcomes_matrices[comp_name][scale] = final_outcomes_matrix
    return final_outcomes_matrices

def plot_multipanel_sensitivity_analysis(
    N2_N1_ratios,
    V_A2_V_A1_ratios,
    final_outcomes_matrices,
    scale_factors,
    competition_types
):
    """
    Plots a 3x3 multipanel heatmap of sensitivity analysis outcomes.

    Parameters:
    - N2_N1_ratios (array): Ratios of initial populations (fast-ev./slow-ev.).
    - V_A2_V_A1_ratios (array): Ratios of evolvability (V_A2/V_A1).
    - final_outcomes_matrices (dict): Nested dictionary with outcomes for each competition type and scaling factor.
    - scale_factors (list): List of scaling factors for V_A1 and V_A2.
    - competition_types (list): List of dictionaries defining competition types and their alpha parameters.
    """
    # Define a universal color_map for all heatmaps
    color_map = {
        "sp 2 (high-genetic-var) wins, sp 1 (low-genetic-var) extinct": 0,
        "sp 1 (low-genetic-var) wins, sp 2 (high-genetic-var) extinct": 1,
        "both extinct": 2,
        "both coexist": 3
    }

    # Create a subplot figure with 3 rows and 3 columns
    fig = make_subplots(
        rows=3, cols=3,
        subplot_titles=[
            f"{comp['name']} | V_A Scale: {scale}"
            for comp in competition_types
            for scale in scale_factors
        ],
        horizontal_spacing=0.1,
        vertical_spacing=0.07
    )

    # Iterate through competition types and scaling factors to plot each heatmap
    for row_idx, competition in enumerate(competition_types, start=1):
        comp_name = competition['name']
        for col_idx, scale in enumerate(scale_factors, start=1):
            final_outcomes_matrix = final_outcomes_matrices[comp_name][scale]
            plot_2d_sensitivity_analysis_initial_pop_vs_evolvability(
                N2_N1_ratios,
                V_A2_V_A1_ratios,
                final_outcomes_matrix,
                subplot_row=row_idx,
                subplot_col=col_idx,
                fig=fig
            )

    # Add a single color scale to the right of all subplots
    # Create a separate colorbar
    fig.add_trace(
        go.Heatmap(
            z=[0, 1, 2, 3],
            colorscale=[
                [0.0, 'red'],       # sp 2 wins, sp 1 extinct
                [1/3, 'blue'],      # sp 1 wins, sp 2 extinct
                [2/3, 'black'],     # both coexist
                [1.0, 'green']      # both extinct
            ],
            showscale=True,
            colorbar=dict(
                title="Outcome",
                tickvals=[0, 1, 2, 3],
                ticktext=list(color_map.keys()),
                len=0.8,
                y=0.5,
                title_side="right",
                titleside="right"
            ),
            zmin=0,
            zmax=3,
            visible=False  # Hide the heatmap itself
        ),
        row=1, col=3  # Position the colorbar on the last subplot
    )

    # Update the overall layout of the figure
    fig.update_layout(
        title="Sensitivity Analysis: Initial Population vs Evolvability Across Competition Types",
        width=1500,  # Adjusted width for better spacing
        height=1200,  # Adjusted height for better spacing
        template='plotly',
        showlegend=False
    )

    # Show the figure
    pio.show(fig)

def main():
    """
    Main function to execute the sensitivity analysis and plotting.
    """
    # Define competition types with their respective alpha parameters
    competition_types = [
        {
            'name': 'Unstable',
            'params': {
                'alpha11': 0.0025,
                'alpha12': 0.005,
                'alpha21': 0.005,
                'alpha22': 0.0025
            }
        },
        {
            'name': 'Neutral',
            'params': {
                'alpha11': 0.00375,
                'alpha12': 0.00375,
                'alpha21': 0.00375,
                'alpha22': 0.00375
            }
        },
        {
            'name': 'Stable',
            'params': {
                'alpha11': 0.005,
                'alpha12': 0.0025,
                'alpha21': 0.0025,
                'alpha22': 0.005
            }
        }
    ]

    # Define the power ranges for the ratios
    N2_N1_powers = np.arange(-3, 3, 0.0625)
    V_A2_V_A1_powers = np.arange(0, 3, 0.0625)

    # Calculate the actual ratios
    N2_N1_ratios = np.power(10.0, N2_N1_powers)
    V_A2_V_A1_ratios = np.power(10.0, V_A2_V_A1_powers)

    # Define the scaling factors for V_A1 and V_A2
    scale_factors = [0.1, 1, 10]

    # Run sensitivity analysis for each competition type and scaling factor
    final_outcomes_matrices = run_multipanel_sensitivity_analysis(
        N2_N1_ratios,
        V_A2_V_A1_ratios,
        scale_factors,
        competition_types,
        base_parameters
    )

    # Plot the multipanel heatmaps
    plot_multipanel_sensitivity_analysis(
        N2_N1_ratios,
        V_A2_V_A1_ratios,
        final_outcomes_matrices,
        scale_factors,
        competition_types
    )

if __name__ == "__main__":
    print('og base', base_parameters.__dict__)
    main()
```

```python
def plot_2d_sensitivity_analysis_initial_pop_vs_evolvability(N2_N1_ratios, V_A2_V_A1_ratios, final_outcomes_matrix):
    color_map = {
        "sp 2 (high-genetic-var) wins, sp 1 (low-genetic-var) extinct": 0,
        "sp 1 (low-genetic-var) wins, sp 2 (high-genetic-var) extinct": 1,
        "both extinct": 2,
        "both coexist": 3
    }

    reverse_color_map = {v: k for k, v in color_map.items()}

    # Convert the final outcomes to numerical values
    numerical_matrix = np.vectorize(color_map.get)(final_outcomes_matrix)

    # Create the heatmap
    heatmap = go.Heatmap(
        z=numerical_matrix.T,  # Transpose the matrix to match the correct axis
        x=N2_N1_ratios,
        y=V_A2_V_A1_ratios,
        colorscale=[[0, 'red'], [1/3, 'blue'], [2/3, 'black'], [1, 'green']],
        showscale = False,
        colorbar=dict(
            tickvals=[0, 1, 2, 3],
            ticktext=list(color_map.keys()),
        )
    )

    fig = go.Figure(data=[heatmap])

    legend_labels = [
        ("Slow evolver extinct", 'red'),
        ("Fast evolver extinct", 'blue'),
        ('Both coexist', 'green'),
        ('Both extinct', 'black')
    ]

    for label, color in legend_labels:
        fig.add_trace(go.Scatter(
            x=[None], y=[None], mode='markers',
            marker=dict(color=color),
            showlegend=True,
            name=label
        ))



    fig.update_layout(
        xaxis=dict(
            title='Initial Population Ratio (log)(fast-ev./slow-ev.)',
            type='log',
            tickvals=[0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000],
            ticktext=["0.001", "0.01", "0.1", "1", "10", "100", "1000"]
        ),
        yaxis=dict(
            title='Evolvability Ratio [fast-ev./slow-ev.] (log)',
            type='log',
            tickvals=[1, 10, 100, 1000],
            ticktext=['1', '10', '100', '1000']
        ),
        template='plotly',
        width=1000,
        height=800,

 
    )
    

    pio.show(fig)

def generate_parameters_with_varied_parameters(N2_N1_ratios, V_A2_V_A1_ratios):
    parameters_list = []
    for N2_N1 in N2_N1_ratios:
        for V_A2_V_A1 in V_A2_V_A1_ratios:
            parameters_dict = Parameters_for_Analysis(base_parameters=base_parameters, V_A1=0.01, V_A2=0.01).__dict__.copy()
            parameters_dict['initial_population_species_2'] = parameters_dict['initial_population_species_1'] * N2_N1
            parameters_dict['V_A2'] = parameters_dict['V_A1'] * V_A2_V_A1
            parameters_dict['alpha11'] = parameters_dict['alpha12'] = 0.0025
            parameters_dict['alpha21'] = parameters_dict['alpha22'] = 0.005
            parameters = Parameters_for_Analysis(base_parameters=base_parameters, **parameters_dict)
            parameters_list.append(parameters)
    return parameters_list

def run_sensitivity_analysis(parameters_list):
    final_outcomes_list = []
    for parameters in parameters_list:
        final_outcome_instance = run_simulation_and_get_outcome_instance(parameters)
        final_outcomes_list.append(final_outcome_instance.outcome)
    return final_outcomes_list

def main():
    N2_N1_powers = np.arange(-3, 3, 0.0625)
    V_A2_V_A1_powers = np.arange(0, 3, 0.0625)
    
    N2_N1_ratios = np.power(10.0, N2_N1_powers)
    V_A2_V_A1_ratios = np.power(10.0, V_A2_V_A1_powers)
    
    parameters_list = generate_parameters_with_varied_parameters(N2_N1_ratios, V_A2_V_A1_ratios)
    final_outcomes_list = run_sensitivity_analysis(parameters_list)
    
    final_outcomes_matrix = np.array(final_outcomes_list).reshape(len(N2_N1_ratios), len(V_A2_V_A1_ratios))
    
    plot_2d_sensitivity_analysis_initial_pop_vs_evolvability(N2_N1_ratios, V_A2_V_A1_ratios, final_outcomes_matrix)

if __name__ == "__main__":
    main()
```
